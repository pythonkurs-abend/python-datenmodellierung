```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({
    'name': ['Bernd', 'Tanja', 'Maria'],
    'alter': [20, 30, 40],
    'anzahl_kinder': [1, 2, 3],
    'anzahl_haustiere': [2, 1, 3]
});

print(df)
```

```python
# Als HTML Tabelle ausgegeben
df
```

```python
df.plot(kind='scatter', x='anzahl_kinder', y='alter', color='green')
```

```python
df.plot(kind='bar', x='name', y='alter')
```

```python
ax = plt.gca()
df.plot(kind='line', x='name', y='anzahl_kinder', ax=ax)
df.plot(kind='line', x='name', y='anzahl_haustiere', color='red', ax=ax)

plt.savefig('test.png')
```
